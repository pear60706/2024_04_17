```python
try:
    name = input('請輸入姓名:')
    height = float(input('請輸入身高(cm):'))
    weight = float(input('請輸入體重(kg):'))
    bmi = weight / (height / 100 ) ** 2
    print(f'{name}, 您的BMI: {round(bmi,ndigits=2)}')
except Exception as e:
    print(f'格式錯誤: {e}')
```
